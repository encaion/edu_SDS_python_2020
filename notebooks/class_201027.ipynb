```python
elec = pd.read_csv("elec_load.csv")
elec.head(2)
```

```python
elec.head(2).loc[:, "X7HR":"X15HR"]
```

```python
elec_melt = elec.melt(id_vars = ["YEAR", "MONTH", "DAY"])
elec_melt.head(2)
```

### Q1. 네 번째 다섯 번째 변수명을 각각 HOUR, LOAD로 변경하시오.

```python
elec_melt = elec_melt.rename(columns = {"variable": "HOUR",
                                        "value": "LOAD"})
```

```python
elec_melt.head(2)
```

### Q2. HOUR변수에서 X와 HR을 제거하시오.

```python
elec_melt["HOUR"] = elec_melt["HOUR"].str.replace(pat = "X|HR", repl = "")
elec_melt.head(2)
```

### Q3. HOUR 변수의 속성을 확인하고 정수형이 아닌 경우 정수형으로 변경하시오.

```python
elec_melt["HOUR"].dtype # O <-- object (문자 또는 문자와 숫자가 섞여있는 상태)
```

```python
elec_melt["HOUR"] = elec_melt["HOUR"].astype("int")
elec_melt["HOUR"].dtype
```

```python
elec_melt.head(2)
```

```python
elec_melt_sub = elec_melt.loc[elec_melt["HOUR"].isin([11, 12, 13]), ]
elec_melt_sub.head(2)
```

### Q4. 시간별 전력 부하량의 평균을 elec_load_mean 객체에 데이터프레임 형태로 저장하시오.

```python
elec_melt_mean = elec_melt.groupby("HOUR")[["LOAD"]].mean().reset_index()
elec_melt_mean.head()
```

```python
def nor_minmax(x):
    value = (x - min(x)) / (max(x) - min(x))
    return value
```

```python
nor_minmax(pd.Series([2, 4, 6, 999]))
```
